**Data Cleaning**

Handles Missing Data

```python
# It will give the sum of missing values in every column
marketing_data.isnull().sum()
```

```python
# It will drop the rows with null values
marketing_data.dropna()
```

```python
# It will fill all the missing values with Ali
marketing_data.fillna(value='Ali')
```

Handle Duplicates

```python
# It will sum the duplicated vales in all columns
marketing_data.duplicated().sum()
# we can also check it in specific columns
marketing_data.duplicated(subset = ['ID', 'Year_Birth']).sum()
```

```python
# Drop the rows with duplicated values
marketing_data = marketing_data.drop_duplicates()
```

Handle Outliers

```python
# It will represent the income in a graph, here we can also take any other column
sns.boxplot(x=marketing_data['Income'])
plt.title('Graph Of Income')
```

**Univariate Analysis**           
One column at a time

**For Numerical Columns:**            
Histogram         
Boxplot        
KDE(density plot)

```python
# A graph which tell the number of people according to year of birth,
# bins tell here the number of dicrete interval(no of buckets) in given range
# kde= True, it stand for Kernal Density Interval, it overlays a smooth curved line, it provide a continuous estimate of data distribution
sns.histplot(marketing_data['Year_Birth'], bins=20, kde=True)
plt.title('Graph of Income')
```

```python
# It will reprsent the MntWines in a graph, the dot are the outliers
sns.boxplot(x = marketing_data['MntWines'])
plt.title('Graph of MntWines')
```

**For Categorical Columns:**  
Countplot         
Pie chart

```python
# Countplot -> It will give us the marital status with count in y axis
sns.countplot(x = 'Marital_Status', data = marketing_data)
plt.title('Graph of Marital Status')
```

```python
# Pieplot -> it will give a circle plot and divide it into parts according to values
# autopct(auto percentage) is use to give the number or percentage of values in the plot like here it give the percentage of each part, we can use different percentage values
marketing_data['Kidhome'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Graph of Kidhome')
```

**Bivariate Analysis:**           
Relationship between two columns

Numerical vs Numerical:

```python
# Scatterplot -> will take Year_Birth in x axis and Income in y axis and represent the data by dots
sns.scatterplot(x = 'Year_Birth', y = 'Income', data = marketing_data)
# we can give it title
plt.title('Relationship between Year Birth and Income')
plt.show() # use to show the graph
```

```python
# regplot(Regression plot) will show data trend and confidence
sns.regplot(x = 'Year_Birth', y = 'Income', data = marketing_data)
plt.title('Relationship between Year Birth and Income')
plt.show()
# if we don't want confidence interval we will take a parameter ci = False
# don't want sccatter dots take parameter scatter = False
# Don't want reg line take parameter fit_reg = False
```

Categorical vs Numerical:

```python
sns.boxplot(y = 'Marital_Status', x = 'Income', data = marketing_data)
plt.title('Relationship between Marital Status and Income')
plt.show()
```

```python
sns.violinplot(y = 'Marital_Status', x = 'Income', data = marketing_data)
plt.title('Relationship between Marital Status and Income')
plt.show()
```

```python
sns.barplot(x = 'Marital_Status', y = 'Income', data = marketing_data)
plt.title('Relationship between Marital Status and Income')
plt.show()
```

Categorical vs Categorical:

```python
pd.crosstab(marketing_data['Education'], marketing_data['Marital_Status']).plot(kind='bar')
plt.title('Relationship between Education and Marital Status')
plt.show()
```

```python
sns.heatmap(pd.crosstab(marketing_data['Marital_Status'], marketing_data['Education']), annot=True)
plt.title('Relationship between Marital Status and Education')
plt.show()
# annot = True calculate the value and display it inside the box
```

**Multivariate Analysis**              
Analyze more than two variables(Columns) at once

```python
# It will analyze the whole dataset , corr() will take only numeric values so we will take only numeric columns
specific_columns = marketing_data[['Year_Birth', 'Income', 'Kidhome', 'Teenhome']]
sns.pairplot(specific_columns)
sns.heatmap(specific_columns.corr(), annot = True, cmap = 'coolwarm')
plt.title('Relationship between Year Birth, Income, Kidhome and Teenhome')
plt.show()
# corr() -> correlation is the statistical measure that tells the direction(positive and negative) and strenght between two numerical values
# cmap(Colormap) = coolwarm -> sets the color scheme for the graph
```

**Correlation and Relationships**         
Find how strongly variables are related.

```python
specific_columns.corr() # It will take only Numeric values
sns.heatmap(specific_columns.corr(), annot = True)
plt.title('Relationship between Year Birth, Income, Kidhome and Teenhome')
plt.show()
```

**Distribution Analysis**    
Understand how data values are spread.

```python
# Will generate a hisgram
sns.distplot(marketing_data['Income'])
plt.title('Distribution of Income')
plt.show()
```

```python
# Will represent the data by a line
sns.kdeplot(marketing_data['Income'])
plt.title('Distribution of Income')
plt.show()
```

```python
# Skewness -> meausre the symmetry of the distribution
specific_columns.skew()
```

```python
# Kurtosis -> statistical measure that describe the shape of tail and the peakedness
specific_columns.kurt()
```

**Feature Relationships / Grouping**    
Use groupby for summarized insights.

```python
# It will give the mean of income by Marital_Status
marketing_data.groupby('Marital_Status')['Income'].mean()
```

```python
# It will group the data by Marital_Status and Income then give their size
marketing_data.groupby(['Marital_Status', 'Income']).size()
```
