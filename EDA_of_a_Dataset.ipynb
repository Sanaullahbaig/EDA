**Multivariate Analysis**              
Analyze more than two variables(Columns) at once

```python
# It will analyze the whole dataset , corr() will take only numeric values so we will take only numeric columns
specific_columns = marketing_data[['Year_Birth', 'Income', 'Kidhome', 'Teenhome']]
sns.pairplot(specific_columns)
plt.title('Relationship between Year Birth, Income, Kidhome and Teenhome')
plt.show()
# corr() -> correlation is the statistical measure that tells the direction(positive and negative) and strenght between two numerical values
# cmap(Colormap) = coolwarm -> sets the color scheme for the graph
```

**The Last Code and Output Explanation:**       
here first we take specific columns and then craeted a pairplot, pairplot give us the relation between all columns, means here we have 4 columns it shows the relation of first column with all 4 columns, then 2 column with all 4 columns and so on. at the last we have heatmap of all the columns.

**Correlation and Relationships**         
Find how strongly variables are related.

```python
specific_columns.corr() # It will take only Numeric values
sns.heatmap(specific_columns.corr(), annot = True)
plt.title('Relationship between Year Birth, Income, Kidhome and Teenhome')
plt.show()
```

**The Last Code and Output Explanation:**  
heatmap is explained in the the previous program but here we take 4 columns instead of groups of a column.

```python
# Skewness -> meausre the symmetry of the distribution
specific_columns.skew()
```

**The Last Code and Output Explanation:**     
We have a KDE line. Skewness measures the symmetry of the distribution. If the line is balanced on both sides, it is symmetrical. If it is lopsided (more on one side), then it is asymmetrical.
If the number is close to 0, the data symmetrical.
If the number is positive like +1, +2 , the data asymmetrical and had a tail on right side.
If the number is negative like -1, -2 , the data asymmetrical and had a tail on left side.          
Here we have Year_Birth with value -0.349944 it means it has a small tail on the left side.
Here we have Income with value 6.763487 it means it has a long tail on the right side.
Here we have Kidhome with value 0.635288 it means it has a small tail on the right side.
Here we have Teenhome with value 0.407115 it means it has a small tail on the right side.

```python
# Kurtosis -> statistical measure that describe of tail and the peakedness of the shape
specific_columns.kurt()
```

**The Last Code and Output Explanation:**
Kurtosis: is a statistical measure that tell us the about two thing in shape of smooth curse how peak the center is and how think the tails are.        
if its value is 0, the peak and tails are normal.
if its value is positive its peak is very tall and skinny and tails are very thick.
if its value is negative its peak is flat and wide and tails are thin.       
Here we have Year_Birth with value 0.717464 it's a small positive number that means the data is slightly peaked in the middle and has slightly thicker tails then a mornal.
Here we have Income with value 159.636700. It means the income data has an extremely sharp peak (because almost everyone earns the same low amount) and very heavy tails (because the few people earning super high amounts pull the tail way out).
Here we have Kidhome with value -0.779661. It's a negative number that means the data for "kids at home" is flatter than a normal shape, with fewer extreme values in the tails.
Here we have Teenhome with value -0.986235. This is also a negative number (very close to -1) that means the data for "teens at home" is quite flat and wide with thin tails.
